```python
row = driver.find_element(By.XPATH,'/html/body/div[3]/div[3]/div[5]/div[1]/table[2]/tbody')
column = driver.find_element(By.XPATH,'/html/body/div[3]/div[3]/div[5]/div[1]/table[2]/tbody/tr[1]/td[1]')
print(len(row.text.split('\n')))
```

```python
df = pd.DataFrame()
for i in range(len(row.text.split('\n'))):
    for j in range(9):
        element = driver.find_element(By.XPATH,f'/html/body/div[3]/div[3]/div[5]/div[1]/table[2]/tbody/tr[{i+1}]/td[{j+1}]')
        df.loc[i, j] = element.text
print(df)
```

```python
## adjust colnames and rownames
column_name=['State', 'Population', 'Murders_and_Nonnegligent_Manslaughter(total_deaths)', 'Murders(total_deaths)', 'Gun_Murders(total_deaths)', 'Gun_Ownership(%)', 'Murder_and_Nonnegligent_Manslaughter_Rate(per100,000)', 'Murder_Rate(per100,000)', 'Gun_Murder_Rate(per100,000)']
df.columns = column_name
df.index = df.index + 1
df_wiki = df.replace(["—[a]", "— [a]", "—[b]", "— [b]"], "0")
df_wiki
```

```python
error = None
i = 1
usa_today = pd.DataFrame()
while True:
    print(i, end=" ")
    templist = []
    try:
        ## State
        if i < 37: ## there is a problem when i = 37
            element = driver.find_element(By.XPATH,f'/html/body/div[2]/main/article/div[5]/h3[{i}]')
            templist.append(element.text.split(". ")[1])
        else:
            element = driver.find_element(By.XPATH,f'/html/body/div[2]/main/article/div[5]/h3[{i+1}]')
            templist.append(element.text.split(". ")[1])

        
        ## Firearm deaths per 100,000 people
        element = driver.find_element(By.XPATH,f'/html/body/div[2]/main/article/div[5]/ul[{i}]/li[1]')
        templist.append(element.text.split(" ")[-3])
        ## Firearm deaths
        element = driver.find_element(By.XPATH,f'/html/body/div[2]/main/article/div[5]/ul[{i}]/li[2]')
        arm_death = re.findall(r"\b\d+\b", element.text)
        templist.append(arm_death[1])
        templist.append(arm_death[2])
        ## Violent Crime Rate per 100,000
        element = driver.find_element(By.XPATH,f'/html/body/div[2]/main/article/div[5]/ul[{i}]/li[3]')
        templist.append(element.text.split(" ")[3])
        ## Permission
        element = driver.find_element(By.XPATH,f'/html/body/div[2]/main/article/div[5]/ul[{i}]/li[4]')
        templist.append(element.text.split(": ")[-1])
        ## Poverty rate
        element = driver.find_element(By.XPATH,f'/html/body/div[2]/main/article/div[5]/ul[{i}]/li[5]')
        templist.append(element.text.split(" ")[2])
    
        usa_today = pd.concat([usa_today, pd.DataFrame([templist])], ignore_index=True)
    
        i = i + 1
    except Exception as e:
        error = str(e)

    if error=="list index out of range":
        print(f"{i}Error!!!", driver.find_element(By.XPATH,f'/html/body/div[2]/main/article/div[5]/ul[{i}]/li[1]').text)
        print(error)
        
    elif error:
        break
    else:
        pass
```

```python
## adjust colnames and rownames
usa_today.columns = columns = ["State", "Firearm_deaths_per_100000", "Firearm_deaths_total", "Firearm_deaths_suicide", "Violent_Crime_Rate_per_100000", "Permission", "Poverty_rate"]
usa_today.index = usa_today.index + 1 
usa_today
```
